#### Importaremos Matplotlib

Ahora vamos a graficar nuestros resultados para ello, vamos a utilziar mapplotlib

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
tabla.species == 'Human'
```

```python
humanos = tabla[tabla.species == 'Human']
```

```python
humanos.head()
```

```python
humanos.homeworld.value_counts()
```

```python
humanos.homeworld.value_counts().plot()
```

#### Con el atributo de kind dentro de plot cambia el tipo de grafica

```python
humanos.homeworld.value_counts().plot(kind="bar")
```

```python
humanos.homeworld.value_counts().plot(kind="bar", color="#0003FC")
```

```python
humanos.homeworld.value_counts().plot(kind="barh", color="#0003FC")
```

```python
humanos.homeworld.value_counts().plot(kind="pie")
```

```python
humanos.homeworld.value_counts().plot(kind="pie", colormap="ocean")
```

```python
humanos.homeworld.value_counts().plot(kind="pie", colormap="ocean", figsize=(15,15))
```
